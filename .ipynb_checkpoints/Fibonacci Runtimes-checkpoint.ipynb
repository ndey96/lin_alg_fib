```python
import numpy as np
import time
import matplotlib.pyplot as plt
import sys
# sys.getrecursionlimit()

def naive_recursion(n):
    if n < 2:
        return n
    return naive_recursion(n-1) + naive_recursion(n-2)

def memoized_recursion(n):
    mem = {0: 0, 1:1}
    def _memoized_recursion(n):
        if n in mem:
            return mem[n]
        mem[n] = _memoized_recursion(n-1) + _memoized_recursion(n-2)
        return mem[n]
    return _memoized_recursion(n)

def naive_matrix(n):
    if n < 2:
        return n
    C = np.array([[1, 1],
                  [1, 0]])
    y_n = C
    for _ in range(n - 1):
        y_n = np.dot(y_n, C)
    return y_n[0,1]

def diagonalized_matrix(n):
    if n < 2:
        return n
    C = np.array([[1, 1],
                  [1, 0]])
    eigenvals, eigenvecs = np.linalg.eig(C)
    y_n = eigenvecs @ np.diag(eigenvals ** n) @ eigenvecs.T
    return y_n[0,1]

###################################################

plt.figure(figsize=(10,8))

max_n = 15
n_vals = np.arange(0,max_n)
times = []
for n in n_vals:
    start = time.time()
    naive_recursion(n)
    times.append(time.time() - start)
plt.plot(n_vals, times, label='Naive Recursion', linestyle='None', marker='x', markersize=12, color='black')

algos = [
    ('Memoized Recursion', memoized_recursion, '+'),
    ('Naive Matrix', naive_matrix, '^'),
    ('Diagonalized Matrix', diagonalized_matrix, 'o'),
]
max_n = 200
for name, algo, marker in algos:
    n_vals = np.arange(0,max_n+1, 5)
    times = []
    for n in n_vals:
        start = time.time()
        algo(n)
        times.append(time.time() - start)
    
#     plt.plot(n_vals, times, label=name)
    plt.plot(n_vals, times, label=name, linestyle='None', marker=marker, markersize=12, color='black')

plt.xlabel('n')
plt.ylabel('Time (s)')
plt.legend()
# plt.axvline(55)
# plt.yscale('log')
```
